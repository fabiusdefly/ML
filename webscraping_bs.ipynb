a - Import libraries

```
#Imports
from bs4 import BeautifulSoup
import requests
```

b - creation of variable url_imdb

```
url_imdb = "https://assets-datascientest.s3.eu-west-1.amazonaws.com/IMDb.html"
```

c - Get HTML page in page_imdb and print first 30 lines

```
page_imdb = requests.get(url_imdb)
```

```
bs_imdb = BeautifulSoup(page_imdb.content, "html.parser")
```

```
bs_imdb
```

```
bs_imdb.prettify().splitlines()[0:30]
```

d - Get html element that contain fiilms infos

```
bs_imdb_data = bs_imdb.find_all("tr")
bs_imdb_data
```

e - Get first film's name

```
first_film = bs_imdb_data[1].find_all('td')
first_film[1].find_all('a')[0].text
```

f - Get first film's rating

i - Import pandas

```
import pandas as pd
```

```
films = bs_imdb_data[1:-1]
films
```

j - Using for loop to get films infos

```
arr = []
for film in films:
  title = film.find_all("td", {"class", "titleColumn"})[0].find_all('a')[0].text
  rating = film.find_all("td", {"class", "ratingColumn imdbRating"})[0].find_all('strong')[0].text
  year = re.sub('[()]', '', film.find_all("td", {"class", "titleColumn"})[0].find_all('span')[0].text)

  arr.append({"title" : title, "rating": rating, "year": year})

df = pd.DataFrame(arr)
```
